```python
'''3. Elabore um algoritmo que solicite ao usuário um número real e ao final imprima na tela se o
número informado é maior que dez, se é menor que dez, ou se é igual a dez
'''
numero_real2=float(input("Informe um número real:  "))
if numero_real2>10:
    print("O número informado foi:", numero_real2, "ele é maior que 10")
elif numero_real2<10:
    print("O número informado foi:", numero_real2, "ele é menor que 10")
elif numero_real2==10:
    print("O número informado foi:", numero_real2, "ele é igual a 10")
```

```python
'''5. Elabore um algoritmo que leia um número inteiro e imprima uma das mensagens: é múltiplo
de 3, ou, não é múltiplo de 3
'''
numero_inteiro1=int(input("Informe um número inteiro:  "))
if numero_inteiro1%3==0:
    print("O número informado foi:", numero_inteiro1, "ele é múltiplo de 3")
elif numero_inteiro1%3!=0:
    print("O número informado foi:", numero_inteiro1, "ele não é múltiplo de 3")
```

```python
'''9. Elabore um algoritmo que leia dois números inteiros e realize a adição; caso o resultado
seja maior que 10, imprima o quadrado do resultado, caso contrário, imprima a metade dele
'''
numero_inteiro1=int(input("Informe um número inteiro:  "))
numero_inteiro2=int(input("Informe um número inteiro:  "))
resultado=numero_inteiro1+numero_inteiro2
if resultado>10:
  print("O número informado foi:", numero_inteiro1, "e", numero_inteiro2)
  print("O resultado da soma dos números é:", resultado)
  print("O quadrado do resultado é:", resultado**2)
```

```python
'''10. O sistema de avaliação de determinada disciplina é composto por três provas. A primeira
prova tem peso 2, a segunda tem peso 3 e a terceira tem peso 5. Considerando que a
média para aprovação é 6.0, faça um algoritmo para calcular a média final de um aluno
desta disciplina e dizer se o aluno foi aprovado ou não
'''
numero_inteiro1=float(input("Informe a nota da primeira prova:  "))
numero_inteiro2=float(input("Informe a nota da segunda prova:  "))
numero_inteiro3=float(input("Informe a nota da terceira prova:  "))
if (numero_inteiro1+numero_inteiro2+numero_inteiro3)/3>=6:
  print("O aluno foi aprovado")
else:
  print("O aluno foi reprovado")
```

```python
'''11. Elabore um algoritmo que leia o nome e o peso de duas pessoas e imprima o nome da
pessoa mais pesada
'''
nome1=input("Informe o nome da primeira pessoa:  ")
peso1=float(input("Informe o peso da primeira pessoa:  "))
nome2=input("Informe o nome da segunda pessoa:  ")
peso2=float(input("Informe o peso da segunda pessoa:  "))
if peso1>peso2:
  print("A pessoa mais pesada é:", nome1)
  print("Ela pesa:", peso1, "kg")
elif peso2>peso1:
  print("A pessoa mais pesada é:", nome2)
  print("Ela pesa:", peso2, "kg")
```

```python
'''13. Elabore um algoritmo que leia dois números e imprima qual é maior, qual é menor, ou se
são iguais
'''
numero_inteiro1=int(input("Informe um número inteiro:  "))
numero_inteiro2=int(input("Informe um número inteiro:  "))
if numero_inteiro1>numero_inteiro2:
  print("O número informado foi:", numero_inteiro1, "ele é maior que", numero_inteiro2)
  print("O número informado foi:", numero_inteiro2, "ele é menor que", numero_inteiro1)
elif numero_inteiro1<numero_inteiro2:
  print("O número informado foi:", numero_inteiro1, "ele é menor que", numero_inteiro2)
  print("O número informado foi:", numero_inteiro2, "ele é maior que", numero_inteiro1)
elif numero_inteiro1==numero_inteiro2:
  print("O número informado foi:", numero_inteiro1, "ele é igual a", numero_inteiro2)
  print("O número informado foi:", numero_inteiro2, "ele é igual a", numero_inteiro1)
```
